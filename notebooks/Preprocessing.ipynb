```python
import os
# txt파일 레이블 재매핑

# 원본 파일이 있는 디렉토리
source_directory = "/home/alpaco/lcm/datas/dataset/val/labels"

# 변경된 파일을 저장할 디렉토리 (테스트)
destination_directory = "/home/alpaco/lcm/datas/dataset/val/data/labelss"

# 저장할 디렉토리가 없으면 생성
os.makedirs(destination_directory, exist_ok=True)

# 디렉토리 내 모든 .txt 파일 가져오기
for filename in os.listdir(source_directory):
    if filename.endswith(".txt"):
        source_path = os.path.join(source_directory, filename)
        destination_path = os.path.join(destination_directory, filename)  # 동일한 파일명으로 저장

        # 파일 읽기
        with open(source_path, "r", encoding="utf-8") as file:
            lines = file.readlines()

        # 첫 번째 값을 조건에 따라 변경
        updated_lines = []
        for line in lines:
            parts = line.strip().split(" ")  # 공백을 기준으로 분리
            if parts:  # 빈 줄이 아닌 경우
                if parts[0] == "0":  # ✅ 0번을 1번으로 변경
                    parts[0] = "1"
                updated_lines.append(" ".join(parts))
            else:
                # 빈 줄인 경우 파일명을 출력
                print("빈 줄 발견 in", filename)

        # 변경된 내용을 새로운 위치에 저장 (원본 유지)
        with open(destination_path, "w", encoding="utf-8") as file:
            file.write("\n".join(updated_lines) + "\n")  # 개행 문자 추가

print("모든 .txt 파일의 첫 번째 값을 '0 → 1'로 변경하여 새로운 위치에 저장 완료!")
```

```python
# 이미지 파일에 맞는 라벨 파일이 있는지
# 라벨 파일에 맞는 이미지 파일이 있는지

import os
import shutil

# 원본 이미지 및 라벨 폴더 경로
image_dir = "/home/alpaco/lcm/datas/images"
label_dir = "/home/alpaco/lcm/datas/labels"

# 이동할 트래시 폴더 경로
trash_image_dir = "/home/alpaco/lcm/datas/dataset/images"
trash_label_dir = "/home/alpaco/lcm/datas/dataset/labels"

# 폴더가 없으면 생성
os.makedirs(trash_image_dir, exist_ok=True)
os.makedirs(trash_label_dir, exist_ok=True)

# 이미지 파일 목록 (확장자 제거)
image_files = {os.path.splitext(f)[0] for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png'))}
# 라벨 파일 목록 (확장자 제거)
label_files = {os.path.splitext(f)[0] for f in os.listdir(label_dir) if f.lower().endswith('.txt')}

# 이미지가 있는데 라벨이 없는 경우 (이동 대상)
missing_labels = image_files - label_files
# 라벨이 있는데 이미지가 없는 경우 (이동 대상)
missing_images = label_files - image_files

#레이블 없는 이미지 이동
if missing_labels:
    print(f"레이블 파일이 없는 이미지 {len(missing_labels)}개 이동 중")
    for file in sorted(missing_labels):
        for ext in ['.jpg', '.png']:  # 두 가지 확장자 지원
            image_path = os.path.join(image_dir, file + ext)
            if os.path.exists(image_path):
                dest_path = os.path.join(trash_image_dir, file + ext)
                shutil.move(image_path, dest_path)  # 이미지 파일 이동
                print(f"이동됨: {image_path} → {dest_path}")

# 이미지 없는 레이블 이동
if missing_images:
    print(f"이미지 파일이 없는 레이블 {len(missing_images)}개 이동 중...")
    for file in sorted(missing_images):
        label_path = os.path.join(label_dir, file + ".txt")
        if os.path.exists(label_path):
            dest_path = os.path.join(trash_label_dir, file + ".txt")
            shutil.move(label_path, dest_path)  # 레이블 파일 이동
            print(f"이동됨: {label_path} → {dest_path}")

# 모든 파일이 정상적으로 짝을 이루는지 확인
if not missing_labels and not missing_images:
    print("모든 이미지와 레이블 파일이 정상적으로 짝을 이룹니다.")
```

```python
# 640x640 패딩 색상 (114, 114, 114) 은 회색(Gray)
import os
import cv2
import numpy as np

# ✅ 원본 이미지 및 레이블 경로
image_folder = "/home/alpaco/lcm/datas/images"
label_folder = "/home/alpaco/lcm/datas/labels"

# ✅ 패딩 적용 후 저장할 폴더
output_image_folder = "/home/alpaco/lcm/datas/dataset/images"
output_label_folder = "/home/alpaco/lcm/datas/dataset/labels"

os.makedirs(output_image_folder, exist_ok=True)
os.makedirs(output_label_folder, exist_ok=True)

# ✅ YOLO 형식 바운딩 박스 변환 (비율 유지)
def resize_with_padding(img, target_size=640):
    h, w, c = img.shape
    scale = target_size / max(w, h)  # 큰 변 기준으로 스케일 조정
    new_w, new_h = int(w * scale), int(h * scale)
    
    # ✅ 리사이즈 (비율 유지)
    resized_img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    
    # ✅ 패딩 추가 (회색 배경)
    pad_w = (target_size - new_w) // 2
    pad_h = (target_size - new_h) // 2
    padded_img = cv2.copyMakeBorder(resized_img, pad_h, pad_h, pad_w, pad_w, cv2.BORDER_CONSTANT, value=(114, 114, 114))

    return padded_img, scale, pad_w, pad_h

# ✅ YOLO 바운딩 박스 변환 함수 (비율 유지 + 패딩 적용)
def adjust_bbox(x, y, w, h, orig_w, orig_h, scale, pad_w, pad_h, new_size=640):
    x_new = (x * orig_w * scale + pad_w) / new_size
    y_new = (y * orig_h * scale + pad_h) / new_size
    w_new = (w * orig_w * scale) / new_size
    h_new = (h * orig_h * scale) / new_size
    return x_new, y_new, w_new, h_new

# ✅ 변환된 이미지 개수 카운트
processed_count = 0

# ✅ 이미지 리사이즈 + 패딩 + 바운딩 박스 변환
for image_name in os.listdir(image_folder):
    if image_name.endswith(('.jpg', '.png', '.jpeg')):
        base_name, ext = os.path.splitext(image_name)

        # ✅ 이미지 로드
        img_path = os.path.join(image_folder, image_name)
        img = cv2.imread(img_path)
        orig_h, orig_w = img.shape[:2]

        # ✅ 패딩 적용 후 640x640 변환 (회색 패딩)
        padded_img, scale, pad_w, pad_h = resize_with_padding(img)

        output_img_path = os.path.join(output_image_folder, image_name)
        cv2.imwrite(output_img_path, padded_img)

        # ✅ 바운딩 박스 변환
        label_path = os.path.join(label_folder, base_name + ".txt")
        output_label_path = os.path.join(output_label_folder, base_name + ".txt")

        if os.path.exists(label_path):
            with open(label_path, "r") as f_in, open(output_label_path, "w") as f_out:
                for line in f_in:
                    parts = line.strip().split()
                    try:
                        class_id = int(float(parts[0]))  # 🔥 float 변환 후 int 변환 (오류 방지)
                        x, y, w, h = map(float, parts[1:])
                        
                        # ✅ YOLO 바운딩 박스 변환 (비율 유지 + 패딩 적용)
                        x_new, y_new, w_new, h_new = adjust_bbox(x, y, w, h, orig_w, orig_h, scale, pad_w, pad_h)

                        # ✅ 새로운 좌표로 저장
                        f_out.write(f"{class_id} {x_new:.6f} {y_new:.6f} {w_new:.6f} {h_new:.6f}\n")

                    except ValueError:
                        print(f"⚠️ [경고] '{label_path}' 파일에서 클래스 ID 변환 실패! (값: {parts[0]})")

        # ✅ 변환된 이미지 개수 카운트 증가
        processed_count += 1

# ✅ 결과 출력
print(f"\n✅ 변환된 이미지 개수: {processed_count}개")
print("✅ **패딩 적용 후 640x640 변환 완료! (회색 배경)** 🚀")
```

```python
# 랜덤으로 train 70 , val 20 , test 10 분류
import os
import shutil
import random

# ✅ 원본 이미지 및 레이블 경로
image_folder = "/home/alpaco/lcm/datas/images"
label_folder = "/home/alpaco/lcm/datas/labels"

# ✅ 데이터셋을 저장할 폴더
output_base = "/home/alpaco/lcm/datas/dataset"

# 🛠 train, val, test 폴더 생성
#for split in ["train", "val", "test"]:
    #os.makedirs(os.path.join(output_base, split, "images"), exist_ok=True)
    #os.makedirs(os.path.join(output_base, split, "labels"), exist_ok=True)

# ✅ 이미지 파일 리스트 가져오기 (확장자 필터링)
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# ✅ 레이블 파일 개수와 동기화 (이미지와 동일한 레이블만 포함)
valid_images = []
for img_file in image_files:
    base_name, _ = os.path.splitext(img_file)
    label_path = os.path.join(label_folder, base_name + ".txt")
    if os.path.exists(label_path):
        valid_images.append(img_file)  # ✅ 레이블이 있는 이미지만 추가

# ✅ 데이터 랜덤 셔플
random.shuffle(valid_images)

# ✅ 데이터 개수 계산
total_count = len(valid_images)
train_count = int(total_count * 0.7)
val_count = int(total_count * 0.2)
test_count = total_count - train_count - val_count  # 남은 데이터를 test로

# ✅ 데이터셋 분할
train_files = valid_images[:train_count]
val_files = valid_images[train_count:train_count + val_count]
test_files = valid_images[train_count + val_count:]

# ✅ 분할된 데이터를 이동
def move_files(files, split):
    for file_name in files:
        base_name, ext = os.path.splitext(file_name)

        src_image = os.path.join(image_folder, file_name)
        src_label = os.path.join(label_folder, base_name + ".txt")

        dest_image = os.path.join(output_base, split, "images", file_name)
        dest_label = os.path.join(output_base, split, "labels", base_name + ".txt")

        shutil.move(src_image, dest_image)
        shutil.move(src_label, dest_label)

move_files(train_files, "train")
move_files(val_files, "val")
move_files(test_files, "test")

print("\n✅ **모든 데이터 분할 완료! (Train: 70%, Val: 20%, Test: 10%)** 🚀")
```

```python
import cv2
import albumentations as A
import numpy as np
import matplotlib.pyplot as plt
import os

# 원본 이미지 폴더 경로
image_folder = "/home/alpaco/findgo/dataset_5000a/cellphone/images"  # 이미지가 들어 있는 폴더
output_folder = "/home/alpaco/lcm/datas/dataset"  # 증강된 이미지를 저장할 폴더
os.makedirs(output_folder, exist_ok=True)  # 폴더가 없으면 생성

# 변환 정의
transform = A.Compose([
    A.Perspective(scale=0.2, p=0.5),
    A.RandomResizedCrop(height=300, width=300, scale=(0.3, 0.8), p=0.5),
    A.CoarseDropout(max_holes=20, max_height=50, max_width=50, fill_value=0, p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.6, contrast_limit=0.6, p=0.5),
], p=1.0)

# 폴더 내 모든 이미지 파일 리스트 가져오기
image_files = [f for f in os.listdir(image_folder) if f.endswith((".jpg", ".png", ".jpeg"))]

# 각 이미지에 대해 증강 적용 및 저장
num_augmented = 5  # 한 이미지당 생성할 증강 이미지 개수

for img_name in image_files:
    img_path = os.path.join(image_folder, img_name)  # 이미지 전체 경로
    image = cv2.imread(img_path)

    if image is None:
        print(f"❌ 오류: {img_path} 을(를) 불러오지 못했습니다.")
        continue  # 이미지 파일이 잘못되었으면 넘어감

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # RGB 변환

    # 여러 개의 증강 이미지 생성
    for i in range(num_augmented):
        augmented = transform(image=image)["image"]

        # 저장할 파일명 생성 (예: "13_aug_0.jpg")
        save_name = f"{os.path.splitext(img_name)[0]}_aug_{i}.jpg"
        save_path = os.path.join(output_folder, save_name)

        # 증강된 이미지 저장
        cv2.imwrite(save_path, cv2.cvtColor(augmented, cv2.COLOR_RGB2BGR))

print(f"✅ 모든 증강된 이미지가 {output_folder} 폴더에 저장되었습니다!")

```
